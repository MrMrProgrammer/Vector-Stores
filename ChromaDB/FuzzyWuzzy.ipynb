```python
from fuzzywuzzy import fuzz
import pandas as pd
from tqdm import tqdm
```

```python
data = pd.read_csv("/home/user/Desktop/data - 05.csv", header=None)
```

```python
data.head()
```

```python
names = list(data[1])
```

```python
names
```

```python
product_name = "غذای خشک گربه پروپلن"

product_names = names

max = 0
name = ""

for product in tqdm(product_names):
    similarity_score = fuzz.ratio(product_name, product)

    if similarity_score > max:
      max = similarity_score
      name = product

print(f"{name}: {max}")
```
