```python
import chromadb
import pandas as pd
from tqdm import tqdm
```

```python
def get_dkp_id(url):
    try:
        chunks = url.split("/")

        for chunk in chunks:
            if chunk.startswith("dkp-"):
                section = chunk.split("-")
                return int(section[1])

        return "The entered link is not related to a product."

    except:
        return "An Error has Occurred"
```
